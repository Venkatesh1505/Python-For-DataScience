```python
#plotting
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
x = [1, 4, 8 , 12]
y = [20, 35,61, 90]
plt.plot(x, y)
```

```python
plt.title("Test title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
```

```python
x1 = [2, 7, 8, 12]
y1 = [23, 45, 78, 67]
plt.figure(figsize=(20,5)) # size of the plot
plt.plot(x1, y1, "y--") # style of the line - yellow dashes
plt.plot(x, y, "go") # style of the line - green dots
plt.legend(["first one", "second one"], loc = "lower right") # Adding legends
plt.xticks(ticks = [2, 7, 8, 12], labels = ["p1", "p2", "p3","p4"]) # adding ticks/labels for x values
plt.show()
```

```python
#Subplots
fig, ax = plt.subplots(nrows = 1, ncols=2, figsize = (4, 4), sharey=True)
ax[0].plot(x,y, "go")
ax[1].plot(x1,y1,"y--")
```

## Different types of plots:
1. Line chart
2. Bar plot
3. Histogram(Distribution)
4. Box plots
5. Violin plot
6. Scatter plot
7. Bubble plot

### Line Chart

```python
#import bigmart data
df1 = pd.read_csv("bigmart_data.csv")
df1 = df1.dropna(how="any")
df1.head()
```

```python
df2 = df1.groupby(['Item_Type'])
df2.first()
```

```python
df3 = df2.Item_MRP.mean()
```

```python
x = df3.index.tolist()
y = df3.values.tolist()
x
```

```python
plt.figure(figsize=(20,5))
plt.title("Item Type vs Item MRP")
plt.xlabel("Item Type")
plt.ylabel("MRP")
plt.plot(x,y)
plt.xticks(labels=x, ticks = np.arange(len(x)))
plt.show()
```

### Bar chart

```python
plt.bar(x,y)
```

### Histogram

```python
#histogram
plt.hist(df1['Item_MRP'], bins=20, color='lightblue')
```

### Box plot

```python
#box plot
data = df1[['Item_Outlet_Sales']]
data1 = data.values
reddiamond = dict(markerfacecolor = 'r', marker = 'D')
plt.boxplot(data1, labels = ['Item_Sales'], flierprops = reddiamond)
```

### Scatter plot:

```python
plt.scatter(df1["Item_Weight"][:200], df1["Item_Visibility"][:200])
```

### Bubble plot:

```python
#bubble plot is one kind of scatter plot but it lets you compare 3 variables
plt.scatter(df1["Item_MRP"][:100], df1["Item_Outlet_Sales"][:100], s=df1["Item_Visibility"][:100]*1000)
```
