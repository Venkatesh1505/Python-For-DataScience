```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#set the style of sns
sns.set()
sns.set(style="darkgrid")

import warnings
warnings.filterwarnings("ignore")
```

```python
#read data
df = pd.read_csv("bigmart_data.csv")
df = df.dropna(how="any")
df = df.reset_index(drop=True)

df["Visibility_Scaled"] = df["Item_Visibility"]*100
df.head()
```

```python
#Line plot
sns.lineplot(x='Item_Weight', y='Item_MRP', data=df[:50])
```

```python
#Bar plot
sns.barplot(x='Item_Type',y='Item_MRP', data = df[:5])
```

```python
#Histogram
sns.distplot(df['Item_MRP'])
```

```python
#Boxplot
sns.boxplot(df['Item_Outlet_Sales'], orient="vertical")
```

```python
#Violin plot
sns.violinplot(df['Item_Outlet_Sales'], orient="vertical", color="magenta")
```

```python
# Scatter plot
sns.relplot(x='Item_Weight', y= 'Item_MRP', data = df[:100], kind="scatter")
```

```python
# Hue semantic
sns.relplot(x='Item_Weight', y = 'Item_MRP', hue='Visibility_Scaled', data = df[:100])
```

```python
#Bubble plot
sns.relplot(x='Item_Weight', y = 'Item_MRP', size='Visibility_Scaled', data=df[:100], kind="scatter")
```

```python
#Categorical plot
sns.relplot(x='Item_Weight', y = 'Item_Visibility', hue='Outlet_Size', style='Outlet_Size', col = 'Outlet_Size', data = df[:100])
```

# Categorical plots
## 1. Categorical scatter plots
    ## a. swarm plot
    ## b. Strip plot
## 2. Categorical distribution plots
    ## a. Boxplot
    ## b. Violin plot
## 3. Categorical estimation plots
    ## a. barplot

```python
#strip plot
sns.catplot(x='Outlet_Size', y = 'Item_Outlet_Sales', kind='strip', data=df[:200])
```

```python
# swarm plot
sns.catplot(x="Outlet_Size", y = "Item_Outlet_Sales", kind = "swarm", data = df[:200])
```

```python
# box plot
sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales', kind = 'box', data = df[:200])
```

```python
# violin plot
sns.catplot(x='Outlet_Size', y = 'Item_Outlet_Sales', kind = 'violin', data = df[:200])
```

```python
# bar plot
sns.catplot(x= 'Outlet_Size', y = 'Item_Outlet_Sales', kind='bar', data = df[:200])
```

## Density plot
