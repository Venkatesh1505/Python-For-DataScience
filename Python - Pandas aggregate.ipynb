# Aggregate functions

### 3 methods:
1. groupby
2. crosstab
3. pivot table

```python
#import libraries
import pandas as pd
#read data
df = pd.read_csv("bigmart_data.csv")
df = df.dropna(how="any")
df = df.reset_index(drop=True)
df.head()
```

```python
#groupby - to group base on item type
df1 = df.groupby("Item_Type")
df1.first()
```

```python
df1.Item_MRP.mean()
```

```python
#you can group by multiple columns
df2 = df.groupby(["Item_Type", "Item_Fat_Content"])
df2.first()
```

### Crosstab:

```python
#crosstab
pd.crosstab(df["Outlet_Size"],df["Outlet_Location_Type"], margins=True)
```

### Pivottable:

```python
#pivottable
pd.pivot_table(df, index = ["Outlet_Establishment_Year"], values=["Item_Outlet_Sales"])
```

```python
import numpy as np
pd.pivot_table(df, index=["Outlet_Establishment_Year", "Outlet_Location_Type"], values = ["Item_Outlet_Sales"], aggfunc = [np.mean, np.median, min, max, np.std])
```
