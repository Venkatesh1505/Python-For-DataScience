# Sorting dataframes

```python
# import libraries
import pandas as pd
#read csv
df = pd.read_csv("bigmart_data.csv")
df.head()
df.shape
```

```python
#Remove null values
df.dropna(how="any", inplace=True)
df.shape
```

```python
df.head()
```

```python
#sorting based on values - df.sort_values - based on one column
df1 = df.sort_values(by="Outlet_Establishment_Year")
df1.head()
```

```python
#sorting based on multiple columns
df2 = df.sort_values(by= ["Outlet_Establishment_Year", "Item_Outlet_Sales"])
df2.head()
```

```python
#sorting parameters - ascending and inplace
df.sort_values(by="Item_Outlet_Sales", ascending=False,inplace=True)
df.head()
```

# Merging dataframes

```python
# create dummy data
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
 

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                     'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                     'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
result = pd.concat([df1,df2,df3])
result
```

```python
#You can also index items from each dataframe to identify uniquely
result1 = pd.concat([df1,df2,df3],keys=["x","y","z"])
result1
```

```python
#to get the value of z dataframe
result1.loc['z']
```

```python
# To concat based on columns - Outer join(Union)
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                        'D': ['D2', 'D3', 'D6', 'D7'],
                        'F': ['F2', 'F3', 'F6', 'F7']},
                       index=[2, 3, 6, 7])
result3 = pd.concat([df1, df4], axis=1, sort=False)
result3
```

```python
# Inner join - Intersection
result4 = pd.concat([df1,df4],axis=1, join="inner")
result4
```

```python
# concat and have the index of the preferred dataframe only - join_axes
result5 = pd.concat([df1,df4], axis=1, join_axes=[df1.index])
result5
```

## Merge

```python
#Merging
df_a = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']})

df_b = pd.DataFrame({
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']})

df_c = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]})
```

```python
result6 = pd.merge(df_a,df_c, on="subject_id")
result6
```

```python
# outer join
result7 = pd.merge(df_a, df_b, on="subject_id", how="outer")
result7
```

```python
#inner join
result8 = pd.merge(df_a, df_b, on ="subject_id", how="inner")
result8
```

```python
#right join
result9 = pd.merge(df_a, df_b, on="subject_id", how="right")
result9
```

```python
#left join
result10 = pd.merge(df_a, df_b, on="subject_id", how="left")
result10
```
